```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("ARB_sensor\merged\label_3.csv")

grouped_data = data.groupby('label')
for label, group in grouped_data:
    # Drop the 'timestamp' and 'label' columns
    group = group.drop([ 'label'], axis=1)
    
    # Create directory for the current label
    save_dir = os.path.expanduser(f"ARB_sensor\heatmap_data/label_{label}")
    os.makedirs(save_dir, exist_ok=True)
    
    # Determine the number of heatmaps (same as original chunk-based calculation)
    #1-450
    #3-27
    #4-2
    #5-2
    #6-170
    #7-200
    #8-85
    chunks = 70
    limit = group.shape[0]
    num_heatmaps = (limit + chunks - 1) // chunks  # Ceiling division
    
    for heatmap_num in range(num_heatmaps):
        # Randomly sample up to 10 rows (adjust if fewer rows are available)
        n_samples = min(10, len(group))
        data_chunk = group.sample(n=n_samples, replace=(n_samples > len(group)))
        
        # Create and save the heatmap
        plt.figure(figsize=(8, 8))
        sns.heatmap(data_chunk, cbar=False, xticklabels=False, yticklabels=False)
        plt.savefig(
            os.path.join(save_dir, f'{str(3)}_heatmap_{heatmap_num}.png'), 
            bbox_inches='tight', 
            pad_inches=0
        )
        plt.close()
```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
for i in range(1,5):


    data = pd.read_csv("ARB_sensor\merged\label_"+str(i)+".csv")

    grouped_data = data.groupby('label')
    for label, group in grouped_data:
        # Drop unnecessary columns
        #group = group.drop(['timestamp', 'label'], axis=1)

        # Create directory for the label
        save_dir = os.path.expanduser(f"ARB_sensor\heatmap_data/label_{label}")
        os.makedirs(save_dir, exist_ok=True)

        n_rows = group.shape[0]
        if n_rows < 5:
            print(f"Label {label} skipped - insufficient rows.")
            continue

        max_start = n_rows - 5
        num_chunks = 5000  # Adjust this to control dataset size

        for chunk_num in range(num_chunks):
            # Randomly select chunk start index
            start_idx = np.random.randint(0, max_start + 1)
            data_chunk = group.iloc[start_idx:start_idx+20, :]

            # Generate and save heatmap
            plt.figure(figsize=(8, 8))
            sns.heatmap(data_chunk, cbar=False, xticklabels=False, yticklabels=False)
            plt.savefig(os.path.join(save_dir, f'3_heatmap_{chunk_num}.png'), 
                        bbox_inches='tight', pad_inches=0)
            plt.close()
```
